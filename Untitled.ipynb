```python
import numpy as np
import pandas as pd
import os,pydicom #,stat
```

```python
path = r'./stage1/'
label_csv = pd.read_csv('stage1_labels.csv')
pid_to_label = {}
```

```python
def judge_inorder():
    for i, pid in enumerate(os.listdir(path)):
        if label_csv['id'][i] != pid:
            print("OOOPSSSS; i:{0}, pid:{1}".format(i,pid))
```

```python
def get_part_pid_to_label():
    for pid in os.listdir(path):
        try:
            tmp = label_csv.loc[label_csv['id'] == pid].values
            pid_to_label[pid] = tmp[0][1]
            #pid_to_label.update({pid: tmp['cancer'][0]})
        except:
            pass
pid_to_label
```
